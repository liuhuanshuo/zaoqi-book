## ÂØºÂÖ•‰∏éÈ¢ÑËÆæ

ËôΩÁÑ∂Âú® `pandas` ‰∏≠ÂèØ‰ª•Áõ¥Êé•Ë∞ÉÁî® `matplotliab` ËøõË°åÂèØËßÜÂåñÔºå‰ΩÜÊòØ‰æùÊóßÈúÄË¶ÅËøõË°åÁõ∏ÂÖ≥ËÆæÁΩÆÔºå‰æãÂ¶ÇÂ≠ó‰Ωì„ÄÅÁ≤æÂ∫¶Á≠â„ÄÇ

````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams['font.sans-serif'] = ['Songti SC']
plt.rcParams['axes.unicode_minus']=False
````

## ÈªòËÆ§ÊñπÊ≥ï

ÈáçÊñ∞ÁîüÊàêÊï∞ÊçÆ df4 ÔºåÂπ∂Âà∂‰ΩúÊï£ÁÇπÂõæÔºåXËΩ¥‰∏∫ aÔºåYËΩ¥‰∏∫ b

````python
df4 = pd.DataFrame(np.random.rand(50, 4), columns=["a", "b", "c", "d"])

df4["species"] = pd.Categorical(
    ["setosa"] * 20 + ["versicolor"] * 20 + ["virginica"] * 10
)
````

````python
df4.plot.scatter(x="a", y="b", figsize=(8, 6))
plt.show()
````

## ‰øÆÊîπÊï£ÁÇπÂ§ßÂ∞è

Âú®‰∏ä‰∏ÄÈ¢òÁöÑÂü∫Á°Ä‰∏äÔºåËÆ©Êï£ÁÇπÁöÑÂ§ßÂ∞èÈöèÁùÄÂÄºÂèòÂåñ

````python
df4.plot.scatter(x="a", y="b", figsize=(8, 6), s=df4["c"] * 200)
plt.show()
````

## Ëøõ‰∏ÄÊ≠•‰ºòÂåñ

Âú®‰∏ä‰∏ÄÈ¢òÁöÑÂü∫Á°Ä‰∏äÔºåËøõË°åÂ¶Ç‰∏ãËÆæÁΩÆ

- Ë∞É‰ΩéÈÄèÊòéÂ∫¶

````python
df4.plot.scatter(x="a", y="b", figsize=(8, 6), marker='o',
                 s=100, linewidths=1, alpha=0.8, edgecolors='black')
plt.show()
````

## È¢úËâ≤Ê∏êÂèò

Âú®‰∏ä‰∏ÄÈ¢òÁöÑÂü∫Á°Ä‰∏äÔºåÂ∞ÜÊï£ÁÇπÁöÑÈ¢úËâ≤ËÆæÁΩÆ‰∏∫Ê∏êÂèòËâ≤

````python
df4.plot.scatter(x="a", y="b",c="c", figsize=(8, 6), marker='o',
                 s=100, linewidths=1, alpha=0.8, edgecolors='black')

plt.show()
````

## ÂàÜÁªÑÁªòÂà∂

Â∞Ü ab ÂàÜ‰∏∫‰∏ÄÁªÑÔºåcdÂàÜ‰∏∫‰∏ÄÁªÑÔºåÂà∂‰Ωú‰∏âÊï£ÁÇπÂõæ

````python
ax = df4.plot.scatter(x="a", y="b", color="DarkBlue", label="Group 1", figsize=(8, 6), marker='o',
                 s=80, linewidths=1, alpha=0.8, edgecolors='black')

df4.plot.scatter(x="c", y="d", color="DarkGreen", label="Group 2", ax=ax, figsize=(8, 6), marker='o',
                 s=80, linewidths=1, alpha=0.8, edgecolors='black');
````
